**Aplikasi Memory Card Bagian 2

*   menerapkan kemampuan memproses peristiwa menggunakan PyQt untuk membuat aplikasi Memory Card
*   memproses menekan tombol “Balas” dan memprogram peralihan antara formulir pertanyaan dan hasil



```python
from PyQt5.QtCore import Qt
from PyQt5.QtWidgets import (
        QApplication, QWidget,
        QHBoxLayout, QVBoxLayout,
        QGroupBox, QButtonGroup, QRadioButton,
        QPushButton, QLabel)


app = QApplication([]) # Membuat instance aplikasi

window = QWidget() # Membuat jendela utama
window.setWindowTitle('Memo Card') # Judul jendela

btn_OK = QPushButton('Answer') # Tombol untuk menjawab
lb_Question = QLabel('The most difficult question in the world!') # Label untuk pertanyaan

RadioGroupBox = QGroupBox("Answer options") # Membuat grup untuk opsi jawaban
rbtn_1 = QRadioButton('Option 1') # Opsi jawaban
rbtn_2 = QRadioButton('Option 2') # Opsi jawaban
rbtn_3 = QRadioButton('Option 3') # Opsi jawaban
rbtn_4 = QRadioButton('Option 4') # Opsi jawaban

#======= bagian 2 =======
RadioGroup = QButtonGroup()
RadioGroup.addButton(rbtn_1)
RadioGroup.addButton(rbtn_2)
RadioGroup.addButton(rbtn_3)
RadioGroup.addButton(rbtn_4)
#=========================

#memulai membuat layout untuk opsi jawaban
layout_ans1 = QHBoxLayout() # layout untuk horizontal
layout_ans2 = QVBoxLayout() # layout untuk vertikal
layout_ans3 = QVBoxLayout() # layout untuk vertikal

layout_ans2.addWidget(rbtn_1) # Tombol 1 di kolom kiri
layout_ans2.addWidget(rbtn_2) # Tombol 2 di kolom kiri
layout_ans3.addWidget(rbtn_3) # Tombol 3 di kolom kanan
layout_ans3.addWidget(rbtn_4) # Tombol 4 di kolom kanan

layout_ans1.addLayout(layout_ans2) # kolom yang ada di sebelah kiri
layout_ans1.addLayout(layout_ans3) # kolom yang ada di sebelah kanan

RadioGroupBox.setLayout(layout_ans1) # Sekelompok tombol radio diatur dalam layout_ans1

# mebuat layout untuk hasil jawaban
AnsGroupBox = QGroupBox("Test result") # Membuat grup widget untuk hasil tes
lb_Result = QLabel('Are you correct or not?') # Tulisan hasil tes
lb_Correct = QLabel('the answer will be here!') # Tulisan jawaban benar

layout_res = QVBoxLayout() # layout vertikal untuk hasil tes
layout_res.addWidget(lb_Result, alignment=(Qt.AlignLeft | Qt.AlignTop)) # menempatkan lb_Result di pojok kiri atas
layout_res.addWidget(lb_Correct, alignment=Qt.AlignHCenter, stretch=2) # menempatkan lb_Correct di tengah dengan stretch 2
AnsGroupBox.setLayout(layout_res) # Menetapkan layout ke grup widget

layout_line1 = QHBoxLayout() # Layout untuk horizontal pertama
layout_line2 = QHBoxLayout() # Layout untuk horizontal kedua
layout_line3 = QHBoxLayout() # Layout untuk horizontal ketiga

layout_line1.addWidget(lb_Question, alignment=(Qt.AlignHCenter | Qt.AlignVCenter)) # menempatkan lb_Question di tengah
layout_line2.addWidget(RadioGroupBox) # menempatkan RadioGroupBox
layout_line2.addWidget(AnsGroupBox)   # menempatkan AnsGroupBox
#RadioGroupBox.hide() # menyembunyikan RadioGroupBox

layout_line3.addStretch(1) # menambahkan stretchable space di kiri tombol
layout_line3.addWidget(btn_OK, stretch=2) # menempatkan btn_OK dengan stretch 2
layout_line3.addStretch(1) # menambahkan stretchable space di kanan tombol

layout_card = QVBoxLayout() # Layout untuk vertikal utama

layout_card.addLayout(layout_line1, stretch=2) # menempatkan layout_line1 dengan stretch 2
layout_card.addLayout(layout_line2, stretch=8) # menempatkan layout_line2 dengan stretch 8
layout_card.addStretch(1) # menambahkan stretchable space
layout_card.addLayout(layout_line3, stretch=1) # menempatkan layout_line3 dengan stretch 1
layout_card.addStretch(1) # menambahkan stretchable space
layout_card.setSpacing(5) # mengatur jarak antar elemen dalam layout utama

#======== Bagian 2 ========
def show_result(): # fungsi untuk menampilkan hasil jawaban
    RadioGroupBox.hide() # menyembunyikan RadioGroupBox
    AnsGroupBox.show() # menampilkan AnsGroupBox
    btn_OK.setText('Next question') # mengubah teks tombol menjadi 'Next question'

def show_question(): # fungsi untuk menampilkan pertanyaan
    RadioGroupBox.show() # menampilkan RadioGroupBox
    AnsGroupBox.hide() # menyembunyikan AnsGroupBox
    btn_OK.setText('Answer') # mengubah teks tombol menjadi 'Answer'
    RadioGroup.setExclusive(False) # menghilangkan batasan sehingga semua radio button bisa dipilih
    rbtn_1.setChecked(False) # menghilangkan pilihan pada semua radio button
    rbtn_2.setChecked(False) # menghilangkan pilihan pada semua radio button
    rbtn_3.setChecked(False) # menghilangkan pilihan pada semua radio button
    rbtn_4.setChecked(False) # menghilangkan pilihan pada semua radio button
    RadioGroup.setExclusive(True) # mengembalikan batasan sehingga hanya satu radio button yang bisa dipilih

def test(): # fungsi untuk menguji jawaban
    if 'Answer' == btn_OK.text(): # jika teks tombol adalah 'Answer'
        show_result() # menampilkan hasil jawaban
    else: # jika teks tombol bukan 'Answer'
        show_question() # menampilkan pertanyaan
#=========================

window = QWidget() # Membuat jendela utama
window.setLayout(layout_card) # Menetapkan layout utama ke jendela
window.setWindowTitle('Memory Card') # Judul jendela

#======== Bagian 2 ========
btn_OK.clicked.connect(test) # Menghubungkan tombol dengan fungsi test()
#=========================

window.show() # Menampilkan jendela
app.exec() # Menjalankan aplikasi

```
